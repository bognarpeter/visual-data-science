```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data_path = 'data/sfbay_final.csv'
df = pd.read_csv(data_path, delimiter=",")
df = df.drop(['Unnamed: 0', 'mercator_y', 'mercator_x'], axis=1)

corrMatrix = df.corr()
fig, ax = plt.subplots(figsize=(10,10))
sn.heatmap(corrMatrix, annot=False,  linewidths=.5, ax=ax)


plt.show()
```

```python
# Example of the Pearson's Correlation test
from scipy.stats import pearsonr

chlorophyll = df['Calculated.Chlorophyll'].fillna(0).tolist()
fluorescene = df['Fluorescence'].fillna(0).tolist()

# print(len(chlorophyll))
# print(len(fluorescene))

stat, p = pearsonr(chlorophyll, fluorescene)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```

```python
test_cmdf = df['Temperature'].dropna()
plt.figure(figsize=(25,7))
plt.title("Temperature change by time")
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.plot(test_cmdf)
```

```python
# Example of the Augmented Dickey-Fuller unit root test
from statsmodels.tsa.stattools import adfuller

data = df['Temperature'].dropna()
stat, p, lags, obs, crit, t = adfuller(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably not Stationary')
else:
	print('Probably Stationary')
```

```python
df_1st_decade = df[
        (pd.to_datetime(df["TimeStamp"]).dt.year >= 1994)
        & (pd.to_datetime(df["TimeStamp"]).dt.year <= 2004)]
df_1st_decade_sal = df_1st_decade["Salinity"].dropna()

df_2nd_decade = df[
        (pd.to_datetime(df["TimeStamp"]).dt.year > 2004)
        & (pd.to_datetime(df["TimeStamp"]).dt.year <= 2014)]
df_2nd_decade_sal = df_2nd_decade["Salinity"].dropna()

plt.figure(figsize=(25,7))
plt.title("Salinity change by time (1994 - 2004)")
plt.xlabel('Time')
plt.ylabel('Salinity')
plt.plot(df_1st_decade_sal)

plt.figure(figsize=(25,7))
plt.title("Salinity change by time (2004 - 2014)")
plt.xlabel('Time')
plt.ylabel('Salinity')
plt.plot(df_2nd_decade_sal)
```

```python
from scipy.stats import ttest_rel

len_df_1 = len(df_1st_decade_sal)
len_df_2 = len(df_2nd_decade_sal)

min_len = len_df_1
if min_len > len_df_2:
    min_len = len_df_2
    
df_1st_decade_sal = df_1st_decade_sal[: min_len]
df_2nd_decade_sal = df_2nd_decade_sal[: min_len]

stat, p = ttest_rel(df_1st_decade_sal, df_2nd_decade_sal)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```
